```python
#3. Write a Python program to calculate Information Gain 𝐼𝐺(𝑆,𝐺𝐸𝑁𝐷𝐸𝑅)for the following 
#dataset used in a Naive Bayes classification model. 
import math
gender = ['Male','Female','Female','Male','Male','Female','Male','Female']
approved = ['Yes','No','Yes','Yes','No','Yes','No','Yes']
def entropy(yes, no):
    total = yes + no
    if yes == 0 or no == 0:
        return 0
    p_yes = yes / total
    p_no = no / total
    return -(p_yes * math.log2(p_yes) + p_no * math.log2(p_no))
total_yes = approved.count('Yes')
total_no = approved.count('No')
entropy_S = entropy(total_yes, total_no)
male_yes = 2
male_no = 2
entropy_male = entropy(male_yes, male_no)
female_yes = 3
female_no = 1
entropy_female = entropy(female_yes, female_no)
total = len(approved)
weighted_entropy = (4/total)*entropy_male + (4/total)*entropy_female
IG = entropy_S - weighted_entropy
print("Entropy(S):", entropy_S)
print("Information Gain IG(S, GENDER):", IG)

```

```python
import numpy as np
X = np.array([45, 23, 27, 41, 56, 60]).reshape(-1, 1)
y = np.array([1, 0, 0, 0, 1, 1])
w = 0.0     # weight
b = 0.0     # bias
learning_rate = 0.001
epochs = 10000
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
n = len(X)

for i in range(epochs):
     z = w * X.flatten() + b
     y_hat = sigmoid(z)
     dw = (1 / n) * np.dot(X.flatten(), (y_hat - y))
     db = (1 / n) * np.sum(y_hat - y)
     w = w - learning_rate * dw
     b = b - learning_rate * db


age = 55
z = w * age + b
probability = sigmoid(z)

if probability >= 0.5:
    print("Predicted BP: High")
else:
    print("Predicted BP: Normal (Low)")
    
print("Weight:", w)
print("Bias:", b)
print("Probability of High BP:", probability)
```
